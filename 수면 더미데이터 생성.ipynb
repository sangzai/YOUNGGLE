#  수면 시작 시간과 수면 종료 시간 생성

```python
from datetime import datetime, timedelta
import random

# 랜덤한 수면 시작 및 종료 시간을 생성하는 함수
def generate_sleep_times(previous_end_time=None):
    start_date = datetime.strptime('2023-11-26', '%Y-%m-%d')
    # end_date = datetime.strptime('2023-12-29', '%Y-%m-%d')

    if previous_end_time is None:
        previous_end_time = start_date

    # 랜덤한 수면 시작 시간 생성
    sleep_start_time = previous_end_time + timedelta(minutes=random.randint(1, 1440))

    # 랜덤한 수면 종료 시간 생성
    sleep_end_time = sleep_start_time + timedelta(minutes=random.randint(60, 1440))

    # 수면 시작 시간이 이전 종료 시간 이후인지 확인
    sleep_start_time = max(sleep_start_time, previous_end_time)

    # 수면 시작 및 종료 시간의 차이가 24시간을 초과하지 않도록 보장
    while (sleep_end_time - sleep_start_time).total_seconds() > 86400:
        sleep_end_time = sleep_start_time + timedelta(minutes=random.randint(60, 1440))

    return sleep_start_time, sleep_end_time

# 주어진 날짜 범위에 대한 수면 시간 생성
sleep_times_list = []
previous_end_time = None
start_num = 1 # Sleepnum 조정

for _ in range(37):  # 원하는 간격 수에 맞게 범위 조정
    sleep_start, sleep_end = generate_sleep_times(previous_end_time)
    sleep_times_list.append([_+start_num, sleep_start.strftime('%Y-%m-%d %H:%M:%S'), sleep_end.strftime('%Y-%m-%d %H:%M:%S')])
    previous_end_time = sleep_end

# 생성된 수면 시간 출력
for entry in sleep_times_list:
    print(entry)
```

# 생성된 수면시간를 조각내주고 수면 심도를 추가해주는 함수

```python
from datetime import datetime, timedelta
import random

ex_list = [6, '2023-12-23 14:41:00', '2023-12-24 02:49:00']

def make_sleep_depth(lst):
    # 수면 심도 설정
    sleep_depth = [4, 5, 7, 8]
    
    # 최소 5회에서 최대 12회까지 수면 분할
    random_number = random.randint(5, 12)
    
    # 시작 및 종료 시간을 문자열에서 datetime 객체로 변환
    start_time = datetime.strptime(lst[1], '%Y-%m-%d %H:%M:%S')
    end_time = datetime.strptime(lst[2], '%Y-%m-%d %H:%M:%S')
    
    # 총 수면 기간 계산
    time_difference = end_time - start_time
    
    # 수면 기간 내에서 무작위 시간 생성
    random_times = [start_time + timedelta(seconds=random.uniform(0, time_difference.total_seconds())) for _ in range(random_number)]
    
    # 날짜 및 시간별로 무작위 시간을 정렬
    random_times.sort()
    
    # 무작위 시간에서 시작 및 종료 시간을 추출하여 수면 세그먼트 리스트 생성
    sleep_segments = [[lst[0], start_time, random_times[0], random.choice(sleep_depth)]]

    # Ensure consecutive segments have different depth values
    for start, end in zip(random_times[:-1], random_times[1:]):
        depth = random.choice(sleep_depth)
        while depth == sleep_segments[-1][-1]:
            # Keep selecting a new depth until it's different from the previous one
            depth = random.choice(sleep_depth)
        sleep_segments.append([lst[0], start, end, depth])

    sleep_segments += [[lst[0], random_times[-1], end_time, random.choice(sleep_depth)]]

    # datetime 객체를 문자열로 형식화
    sleep_segments_formatted = [[sleep_num, start.strftime('%Y-%m-%d %H:%M:%S'), end.strftime('%Y-%m-%d %H:%M:%S'), depth] for sleep_num, start, end, depth in sleep_segments]

    return sleep_segments_formatted

result = make_sleep_depth(ex_list)
for value in result:
    print(value)
```

# 조각 내주는 함수 적용

```python
data = []
for entry in sleep_times_list:
  result = make_sleep_depth(entry)
  for value in result:
    print(value)
    data.append(value)
```

# 1회 수면을 랜덤하게 조각 내고 자세를 추가해주는 함수

```python
from datetime import datetime, timedelta
import random

ex_list = [6, '2023-12-23 14:41:00', '2023-12-24 02:49:00']

def make_sleep_posture(lst):
    # 수면 심도 설정
    posture = ['DP','CP']
    
    # 최소 5회에서 최대 12회까지 수면 분할
    random_number = random.randint(2, 7)
    
    # 시작 및 종료 시간을 문자열에서 datetime 객체로 변환
    start_time = datetime.strptime(lst[1], '%Y-%m-%d %H:%M:%S')
    end_time = datetime.strptime(lst[2], '%Y-%m-%d %H:%M:%S')
    
    # 총 수면 기간 계산
    time_difference = end_time - start_time
    
    # 수면 기간 내에서 무작위 시간 생성
    random_times = [start_time + timedelta(seconds=random.uniform(0, time_difference.total_seconds())) for _ in range(random_number)]
    
    # 날짜 및 시간별로 무작위 시간을 정렬
    random_times.sort()
    
    # 무작위 시간에서 시작 및 종료 시간을 추출하여 수면 세그먼트 리스트 생성
    sleep_segments = [
        [ 
         random.choice(posture),
         start_time, 
         random_times[0]
        ]
    ]

    # Ensure consecutive segments have different depth values
    for start, end in zip(random_times[:-1], random_times[1:]):
        posture = 'CP' if sleep_segments[-1][0] == 'DP' else 'DP'
        sleep_segments.append([ posture, start, end ])

    sleep_segments += [
        [ 
         'CP' if sleep_segments[-1][0] == 'DP' else 'DP',
         random_times[-1], 
         end_time
         ]
        
        ]

    # datetime 객체를 문자열로 형식화
    sleep_segments_formatted = [
        [ 
         'test',
         posture,
         start.strftime('%Y-%m-%d %H:%M:%S'), 
         end.strftime('%Y-%m-%d %H:%M:%S'), 
         ] for posture, start, end  in sleep_segments
        ]

    return sleep_segments_formatted

result = make_sleep_posture(ex_list)
for value in result:
    print(value)
```
